## **Generate text from image and text inputs**

```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

load_dotenv()
api_key=os.getenv("GOOGLE_API_KEY")
# Create a client
genai.configure(api_key=api_key)
```

```python
import PIL.Image

img = PIL.Image.open('image1.jpg')
img


```

```python
model = genai.GenerativeModel('gemini-pro-vision')
response = model.generate_content(img)

print(response.text)
```

```python
model = genai.GenerativeModel('gemini-pro-vision')
response = model.generate_content(["Write a short, engaging blog post based on this picture.", img])
print(response.text)
```

## Fountain of Mirrors

Have you ever seen a fountain of mirrors? It's a truly mesmerizing sight. The way the mirrors reflect the water and the sky creates an infinite number of images. It's like looking into a portal to another world.

This particular fountain is located in the city of Lyon, France. It was created by the artist Yaacov Agam in 1975. The fountain is made up of 12 stainless steel spheres, each of which is 2 meters in diameter. The spheres are arranged in a circle, and they reflect the surrounding buildings and sky.

The fountain is a popular tourist attraction, and it's easy to see why. It's a beautiful and unique work of art, and it's a great place to relax and enjoy the scenery.

If you're ever in Lyon, be sure to check out the Fountain of Mirrors. It's a truly unforgettable experience.
